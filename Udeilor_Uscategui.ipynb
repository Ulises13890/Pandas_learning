Descarga del dataset desde kagglehub

```python
import pandas as pd
import kagglehub

path = kagglehub.dataset_download("wuxi5791/fruitcsv")

print("Path to dataset files:", path)
```

Conversion del dataset en dataframe con pandas

```python
import os 
# Definir el path original 
# Agregar "\fruit.csv" al path 
new_path = os.path.join(path, 'fruit.csv')
df_fruit=pd.DataFrame(pd.read_csv(new_path))
print(df_fruit)
```

```python
# Crear una nueva columna 'fruit_name_nuevo' basándose en 'fruit_label'
df_fruit['fruit_name_nuevo'] = df_fruit['fruit_label'].map({1: 'apple', 2: 'mandarin', 3: 'orange', 4: 'lemon'})

# Reemplazar la columna 'fruit_name' con la nueva columna
df_fruit['fruit_name'] = df_fruit['fruit_name_nuevo']

# Eliminar la columna auxiliar
df_fruit.drop('fruit_name_nuevo', axis=1, inplace=True)

print(df_fruit)
```

```python
df_fruit.fruit_name.unique()
```

Vamos a hacer limpieza de los valores de fruit_name

```python
# Crear una nueva columna 'fruit_name_nuevo' basándose en 'fruit_name'
df_fruit['fruit_name_nuevo'] = df_fruit['fruit_name'].map({'apple': 'apple',' apple': 'apple','  apple' : 'apple','mandarin': 'mandarin','mandarin ': 'mandarin', 'orange': 'orange', 'lemon': 'lemon'})

# Reemplazar la columna 'fruit_name' con la nueva columna
df_fruit['fruit_name'] = df_fruit['fruit_name_nuevo']

# Eliminar la columna auxiliar
df_fruit.drop('fruit_name_nuevo', axis=1, inplace=True)

print(df_fruit)
```

```python
df_fruit.fruit_name.unique()
```

Contamos la cantidad de veces que se repite fruit_name y lo mostramos como un dataframe

```python
# Suponiendo que tu DataFrame se llama 'df_fruit'
conteo_frutas = df_fruit['fruit_name'].value_counts()

# Convertir la serie a un DataFrame
df_conteo = conteo_frutas.to_frame(name='cantidad')

# Transponer el DataFrame para que los nombres de las frutas sean columnas
df_conteo = df_conteo.transpose()

print(df_conteo)
```
